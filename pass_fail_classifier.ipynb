```python

import pandas as pd
from sklearn.linear_model import LogisticRegression

# Create dataset
data = {
    "Study Hours": [1, 2, 3, 4, 5, 6, 7],
    "Pass": [0, 0, 0, 1, 1, 1, 1]
}

df = pd.DataFrame(data)

# Features and target
X = df[["Study Hours"]]
y = df["Pass"]

# Train model
model = LogisticRegression()
model.fit(X, y)

# Predict for 6 hours
new_data = pd.DataFrame([[6]], columns=["Study Hours"])
prediction = model.predict(new_data)

print("Prediction for 6 hours:", prediction)
```
