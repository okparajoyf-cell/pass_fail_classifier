```python

import pandas as pd
from sklearn.linear_model import LogisticRegression

# Data: study hours vs pass (1) or fail (0)
data = {
    "Study Hours": [1, 2, 3, 4, 5, 6, 7],
    "Pass": [0, 0, 0, 1, 1, 1, 1]
}

df = pd.DataFrame(data)

X = df[["Study Hours"]]
y = df["Pass"]

# Create model
model = LogisticRegression()
model.fit(X, y)

# Predict
prediction = model.predict([[3]])
print("Prediction for 3 hours:", prediction)
```
